# Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Wrangling

## Gethering Data

```python
customer_df = pd.read_csv("customers.csv")
customer_df
```

```python
orders_df = pd.read_csv("orders.csv")
orders_df
```

```python
products_df = pd.read_csv("products.csv")
products_df
```

```python
sales_df = pd.read_csv("sales.csv")
sales_df
```

## Assessing Data

```python
customer_df.info()
customer_df.isnull().sum()
customer_df.duplicated().sum()
customer_df.describe()
```

```python
orders_df.info() # order_date dan delivery_date diganti dengan type data date time
orders_df.isnull().sum()
orders_df.duplicated().sum()
orders_df
```

```python
products_df.info() 
products_df.isnull().sum()
products_df.duplicated().sum()
```

```python
sales_df.info() 
sales_df.isnull().sum()
sales_df.duplicated().sum()
```

## Cleaning Data

```python
orders_df.info()
datetime_columns = ['order_date','delivery_date']
for column in datetime_columns:
    orders_df[column]=pd.to_datetime(orders_df[column])
orders_df.info()
orders_df
```

# Exploratory Data Alalysis 

```python
customer_id_in_orders_df = orders_df.customer_id.tolist()
customer_df['status'] = customer_df['customer_id'].apply(lambda x: 'Active' if x in customer_id_in_orders_df else 'Non Active')
customer_df.head()
```

```python
customer_df.groupby(by='status').customer_id.nunique().sort_values(ascending=False)
```

```python
all_df= pd.merge(
    left = sales_df,
    right =  orders_df,
    how='left',
    left_on='order_id',
    right_on='order_id'
)
all_df
```

```python
all_df = pd.merge(
    left = all_df,
    right = products_df,
    how='left',
    left_on='product_id',
    right_on='product_ID'
)
all_df.info()
```

```python
all_df = pd.merge(
    left = all_df,
    right = customer_df,
    how='left',
    left_on='customer_id',
    right_on='customer_id'
)
all_df.info(35)
```

```python
all_df.groupby(by="gender").agg({
    "customer_id": "nunique",
    "age": ["max", "min", "mean", "std"]
})
```

```python
all_df["age_group"] = all_df.age.apply(lambda x: "Youth" if x <= 25 else ("Seniors" if x > 60 else "Adults"))
all_df.groupby(by="age_group").order_id.nunique().sort_values(ascending=False)
```

```python
all_df.groupby(by="state").agg({
    "order_id": "nunique",
    "total_price" : "sum"
}).sort_values(by='total_price',ascending=False)
```

```python
all_df.groupby(by=['order_date']).agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
})
```

```python
all_df.groupby(by=['product_type', 'order_date']).agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
})
```

```python
all_df.groupby(by="product_name").agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
}).sort_values(by="total_price", ascending=False)
```

```python
all_df.groupby(by=['product_type','product_name']).agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
}).sort_values(by=['product_type','total_price'],ascending=False)
```

```python
all_df.groupby(by=['state','product_type']).agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
}).sort_values(by=['state','total_price'],ascending=False)
```

# Visualisasi & Explanatory Analysis

```python
all_df.groupby(by="gender").agg({
    "customer_id": "nunique",
})
# Aggregasi data
aggregated_data = all_df.groupby(by="gender").agg({
    "customer_id": "nunique",
}).reset_index()
max_prop_index = aggregated_data['customer_id'].idxmax()
explode = [0.1 if i == max_prop_index else 0 for i in range(len(aggregated_data))]
# Visualisasi dengan pie chart
plt.figure(figsize=(8, 8))
plt.pie(aggregated_data['customer_id'], labels=aggregated_data['gender'], autopct='%1.1f%%', startangle=90, explode=explode)
plt.title('Proporsi Jumlah Customer berdasarkan Gender')
plt.show()
```

```python
all_df["age_group"] = all_df.age.apply(lambda x: "Youth" if x <= 25 else ("Seniors" if x > 60 else "Adults"))

# Menghitung jumlah unik order_id berdasarkan age_group
age_group_counts = all_df.groupby(by="age_group").order_id.nunique().sort_values(ascending=False)

# Visualisasi dengan pie chart
plt.figure(figsize=(8, 8))
plt.pie(age_group_counts, labels=age_group_counts.index , autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral', 'lightgreen'])
plt.title('Proporsi Jumlah Penjualan berdasarkan Grup Usia')
plt.show()
```

```python
aggregated_data = all_df.groupby(by="state").agg({
    "order_id": "nunique",
    "total_price": "sum"
}).sort_values(by='total_price', ascending=False).reset_index()

# Visualisasi dengan bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='state', y='order_id', data=aggregated_data, palette='deep')
plt.title('Jumlah Penjualan berdasarkan State')
plt.xlabel('State')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Aggregasi data
aggregated_data = all_df.groupby(by="state").agg({
    "order_id": "nunique",
    "total_price": "sum"
}).sort_values(by='total_price', ascending=False).reset_index()

# Visualisasi dengan bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='state', y='total_price', data=aggregated_data, palette='viridis')
plt.title('Total Penjualan berdasarkan State')
plt.xlabel('State')
plt.ylabel('Penjualan')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
all_df['month'] = all_df['order_date'].dt.to_period('M')

aggregated_data = all_df.groupby(by='month').agg({
    'sales_id': 'nunique',
    'quantity_x': 'sum',
    'total_price': 'sum'
}).reset_index()

# Konversi 'month' ke string sebelum plotting
aggregated_data['month'] = aggregated_data['month'].astype(str)

# Visualisasi dengan line plot menggunakan Matplotlib
plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='sales_id', data=aggregated_data, marker='o', color='b')
plt.title('Jumlah Penjualan Berdasarkan Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=45, ha='right')  # Rotate month labels for readability
plt.show()

```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='total_price', data=aggregated_data, marker='o', color='g')
plt.title('Jumlah Barang Terjual Berdasarkan Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Barang Terjual')
plt.xticks(rotation=45, ha='right')  # Rotate month labels for readability
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='quantity_x', data=aggregated_data, marker='o', color='r')
plt.title('Total Omset Berdasarkan Bulan')
plt.xlabel('Bulan')
plt.ylabel('Total Omset')
plt.xticks(rotation=45, ha='right')  # Rotate month labels for readability
plt.show()
```

```python
all_df['month'] = all_df['order_date'].dt.to_period('M')

# Aggregasi data
aggregated_data = all_df.groupby(by=['state', 'product_type']).agg({
    'sales_id': 'nunique',
    'quantity_x': 'sum',
    'total_price': 'sum'
}).sort_values(by=['state', 'total_price'], ascending=False).reset_index()

# Visualisasi dengan bar plot menggunakan Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='state', y='sales_id', hue='product_type', data=aggregated_data, palette='deep')
plt.title('Jumlah Penjualan Berdasarkan State dan Product Type')
plt.xlabel('State')
plt.ylabel('Jumlah Penjualan')
plt.legend(title='Product Type', bbox_to_anchor=(1.05, 1), loc='upper left')  # Menampilkan legenda di luar plot
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='state', y='quantity_x', hue='product_type', data=aggregated_data, palette='muted')
plt.title('Jumlah Barang Terjual Berdasarkan State dan Product Type')
plt.xlabel('State')
plt.ylabel('Jumlah Barang Terjual')
plt.legend(title='Product Type', bbox_to_anchor=(1.05, 1), loc='upper left')  # Menampilkan legenda di luar plot
plt.xticks(rotation=45, ha='right')
plt.show()
```
